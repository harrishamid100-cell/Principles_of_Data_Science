```python
diabetesdata = pd.read_csv('/content/drive/MyDrive/diabetes.csv')
diabetesdata
```

```python
train_data = diabetesdata.sample(frac=0.6, random_state=25)
train_data
```

```python
mean_data = train_data['Glucose'].mean()
print("The average Glucose in this training data is", mean_data.round(2))
```

```python
max_data = train_data['Glucose'].max()
print("The highest Glucose amount in this training data is", max_data)
```

```python
meandata = diabetesdata['Glucose'].mean()
print("The average Glucose in this full data is", meandata.round(2))
```

```python
maxdata = diabetesdata['Glucose'].max()
print("The highest Glucose amount in this training data is", maxdata)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(diabetesdata['Glucose'], kde=True)
plt.axvline(x=meandata, color='green', linestyle='dashed', label='Mean of Glucose (Actual)')
plt.title('Distribution of Mean of Glucose Levels (Actual)')
plt.xlabel('Glucose Level')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(train_data['Glucose'], kde=True)
plt.axvline(x=mean_data, color='green', linestyle='dashed', label='Mean of Glucose (Training Sample)')
plt.title('Distribution of Mean of Glucose Levels (Training Sample)')
plt.xlabel('Glucose Level')
plt.ylabel('Frequency')
plt.legend()
plt.show()

max(diabetesdata['Glucose'])
```

The mean of Glucose in the actual data is much lower than the training sample data of the average Glucose

```python
mean_bloodpressure = diabetesdata['BloodPressure'].mean()
mean_bloodpressure
```

```python
bootstrap_sample = []
for _ in range(500):
    sample = np.random.choice(diabetesdata['BloodPressure'], size=150, replace=True)
    bootstrap_sample.append(sample)

print(f"Generated {len(bootstrap_sample)} bootstrap samples.")
```

```python
bootstrap_mean = np.mean(bootstrap_sample)
bootstrap_mean
```

```python
bootstrap_std = np.std(bootstrap_sample)
bootstrap_std
```

```python
percentileboot = np.percentile(bootstrap_sample, 95)
percentileboot
```

```python
percentile95 = diabetesdata['BloodPressure'].quantile(0.95)
percentile95
```

```python
stddata = diabetesdata['BloodPressure'].std()
stddata
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(diabetesdata['BloodPressure'], kde=True)
plt.axvline(x=percentile95, color='red', linestyle='dashed', label='95th Percentile')
plt.title('Distribution of Blood Pressure with 95th Percentile (Actual)')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(diabetesdata['BloodPressure'], kde=True)
plt.axvline(x=percentileboot, color='red', linestyle='dashed', label='95th Percentile (Bootstrap)')
plt.title('Distribution of Blood Pressure with 95th Percentile (Bootstrap)')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(diabetesdata['BloodPressure'], kde=True)
plt.axvline(x=mean_bloodpressure, color='red', linestyle='dashed', label='Mean of BloodPressure (Actual)')
plt.title('Distribution of Mean Blood Pressure (Actual)')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(diabetesdata['BloodPressure'], kde=True)
plt.axvline(x=bootstrap_mean, color='red', linestyle='dashed', label='Mean of BloodPressure (Bootstrap)')
plt.title('Distribution of Mean Blood Pressure (Bootstrap)')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(diabetesdata['BloodPressure'], kde=True)
plt.axvline(x=bootstrap_std, color='red', linestyle='dashed', label='Standard Deviation of BloodPressure (Bootstrap)')
plt.title('Distribution of Blood Pressure with Standard Deviation (Bootstrap)')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(diabetesdata['BloodPressure'], kde=True)
plt.axvline(x=stddata, color='red', linestyle='dashed', label='Standard Deviation of BloodPressure (Actual)')
plt.title('Distribution of Blood Pressure with Standard Deviation (Actual)')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Using the bootstrap of mean, standard deviation, and 95 percentile of data and comparing it with the actual data, we see that bootstrap displays the exact results as the actual data.
